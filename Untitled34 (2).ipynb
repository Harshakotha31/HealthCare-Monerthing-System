```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, classification_report
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, Dense, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
```

```python
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
possible_files = ['ECG5000_test.csv']
datafile = None
for f in possible_files:
    if os.path.exists(f):
        datafile = f
        break

if datafile is None:
    raise FileNotFoundError("No dataset file found. Upload ECG CSV and set 'datafile' path.")

print("Using data file:", datafile)
df = pd.read_csv(datafile)
```

```python
df.head()
```

```python
if df.shape[1] > 1:
    # heuristic: if last column contains small integers (labels)
    last_col = df.columns[-1]
    if pd.api.types.is_integer_dtype(df[last_col]) or df[last_col].nunique() < 10:
        X = df.drop(columns=[last_col]).values.astype(float)
        y = df[last_col].values
    else:
        # assume no label, single-signal rows
        X = df.values.astype(float)
        y = None
else:
    # single column: maybe a single long time series -> reshape into windows (not automated here)
    raise ValueError("Single-column dataset detected. Please segment into fixed-length windows first.")

print("X shape:", X.shape, "y shape:", None if y is None else y.shape)
```

```python
if y is not None:
    vals, counts = np.unique(y, return_counts=True)
    normal_label = vals[np.argmax(counts)]
    y_binary = (y != normal_label).astype(int)  # 0 = normal, 1 = anomaly
    print("Normal label:", normal_label, "| total normal samples:", (y_binary==0).sum())
else:
    y_binary = None
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
timesteps = X_scaled.shape[1]
X_seq = X_scaled.reshape((X_scaled.shape[0], timesteps, 1))
```

```python
if y_binary is not None:
    X_train = X_seq[y_binary == 0]
    X_test  = X_seq  # we'll evaluate on all
    y_test  = y_binary
else:
    # if no labels, do a train/test random split and later inspect reconstruction errors
    from sklearn.model_selection import train_test_split
    X_train, X_test = train_test_split(X_seq, test_size=0.2, random_state=SEED)
    y_test = None

print("X_train shape:", X_train.shape, "X_test shape:", X_test.shape)
```

```python
latent_dim = 64
input_shape = (timesteps, 1)
inp = Input(shape=input_shape)
```

```python
# Encoder
x = LSTM(128, activation='tanh', return_sequences=True)(inp)
x = LSTM(latent_dim, activation='tanh', return_sequences=False)(x)
```

```python
# Bottleneck -> repeat
x = RepeatVector(timesteps)(x)
```

```python
# Decoder
x = LSTM(latent_dim, activation='tanh', return_sequences=True)(x)
x = LSTM(128, activation='tanh', return_sequences=True)(x)
out = TimeDistributed(Dense(1))(x)

autoencoder = Model(inputs=inp, outputs=out)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

```python
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
history = autoencoder.fit(
    X_train, X_train,
    epochs=100,
    batch_size=64,
    validation_split=0.1,
    callbacks=[es],
    verbose=2
)
```

```python
plt.figure()
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('epoch')
plt.legend()
plt.title('Training loss')
plt.show()
```

```python
X_test_pred = autoencoder.predict(X_test)
mse = np.mean(np.square(X_test_pred - X_test), axis=(1,2))
```

```python
if y_test is not None:
    auc = roc_auc_score(y_test, mse)
    print("ROC-AUC (reconstruction error):", auc)
    # choose threshold such that we capture e.g. 95th percentile of train reconstruction errors
    train_pred = autoencoder.predict(X_train)
    train_mse = np.mean(np.square(train_pred - X_train), axis=(1,2))
    threshold = np.percentile(train_mse, 95)
    print("Threshold set to 95th percentile of train MSE:", threshold)

    y_pred = (mse > threshold).astype(int)
    print(classification_report(y_test, y_pred, digits=4))

else:
    # no labels, show histogram of MSE so you can pick threshold manually
    train_pred = autoencoder.predict(X_train)
    train_mse = np.mean(np.square(train_pred - X_train), axis=(1,2))
    threshold = np.percentile(train_mse, 95)
    print("No true labels. Suggested threshold (95th percentile of train MSE):", threshold)
```

```python
plt.figure()
plt.hist(mse, bins=100)
plt.xlabel('Reconstruction MSE')
plt.title('Test reconstruction error distribution')
plt.axvline(threshold, linestyle='--')
plt.show()
```

```python
anomaly_idxs = np.where(mse > threshold)[0]
normal_idxs = np.where(mse <= threshold)[0]
```

```python
print("Detected anomalies:", len(anomaly_idxs))
```

```python
n_show = 4

plt.figure(figsize=(12, 6))
for i, idx in enumerate(anomaly_idxs[:n_show]):
    plt.subplot(2, n_show, i+1)
    plt.plot(X_test[idx].reshape(-1))
    plt.title(f"Anom idx {idx} | mse={mse[idx]:.4f}")
    # overlay reconstruction
    plt.subplot(2, n_show, i+1+n_show)
    plt.plot(X_test_pred[idx].reshape(-1))
    plt.title("Reconstruction")
plt.tight_layout()
plt.show()
```
